```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# Define State
class State(TypedDict):
    a: int
    b: int
    c: int
```

```python
#Define Node Functon
def addTwoNumber(state: State):
    x = state["a"]
    y = state["b"]
    z = x+y
    state["c"]=z
    return state
```

```python
# Construct Graph Layout
workflow = StateGraph(State)
```

```python
# Add your Defined Node To The Workflow Structure
workflow.add_node("addTwoNumber", addTwoNumber)
```

```python
# Define Edges
workflow.add_edge(START, "addTwoNumber")
workflow.add_edge("addTwoNumber", END)
```

```python
# Complile The Graph
app = workflow.compile()
```

```python
app
```

```python
# Invoke And Run The Graph
initial_input= {"a":3, "b":4}
final_state = app.invoke(initial_input)
```
